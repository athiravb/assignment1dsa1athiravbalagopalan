```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("iris.xls")#reading dataset
```

```python
data.head()
data.columns#to display the columns
```

```python
#finding mean of each column
data.mean(numeric_only=True)
```

```python
#check for the null values present in the dataset.
data.isnull().sum()
```

```python
#plotting bar diagram for SL
data[['SL', 'SW', 'PL','PW']].plot.bar()
plt.title('BAR DIAGRAM',fontsize=20)
plt.show()
```

```python
#plotting boxplot of SL
sns.boxplot(data['SL'])
```

```python
#scatter plot of SL versus PW
plt.scatter(x=data['SL'], y=data['PW'],s=15)
plt.title('Scatter Diagram of SL & PW',color='red',fontsize=20)
plt.show()
```

```python
plt.hist(data['SL'], color = 'violet')
plt.title('HISTOGRAM',fontsize=18)
plt.show()
```
